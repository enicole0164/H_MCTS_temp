```python
total_success = 0
iter_list = []
for random_seed in range(100):
    MCTS = Plain_MCTS_Cont(grid_setting, random_seed=random_seed,
                          num_barrier = 15)
    traj, success, iter = MCTS.search()
    iter_list.append(iter) += 1
    total_success += success
    if success:
        MCTS.env.plot_grid(0, traj)
        print(f'success with {iter}')
    else:
        MCTS.env.plot_grid(0, traj)
        print('Failed')
    
```
